```
import pandas as pd; pd.set_option('display.min_rows', 50); pd.set_option('display.max_rows', 200)
import numpy as np
import yfinance as yf
from google.colab import files
```

```
def busca_carteira_teorica(indice):
  url = 'http://bvmf.bmfbovespa.com.br/indices/ResumoCarteiraTeorica.aspx?Indice={}&idioma=pt-br'.format(indice.upper())
  return pd.read_html(url, index_col='Código', thousands='.', decimal=',')[0][:-1]
```

```
ibov = busca_carteira_teorica('ibov')
```

```
ibov.sort_values('Part. (%)', ascending=False)
```

```
idiv = busca_carteira_teorica('idiv') 
```

```
idiv.sort_values('Part. (%)', ascending=False)
```

```
ibrx = busca_carteira_teorica('ibrx')
```

```
ibrx.sort_values('Part. (%)', ascending=False)
```

```
pd.concat([ibov, ibrx], keys=['IBOV', 'IBRX'], axis=1)
```

```
def get_tickers(indice):
  url = 'http://bvmf.bmfbovespa.com.br/indices/ResumoCarteiraTeorica.aspx?Indice={}&idioma=pt-br'.format(indice.upper())
  return (pd.read_html(url, index_col='Código', thousands='.', decimal=',')[0][:-1].index + '.SA').to_list()
```

```
ibov_tickers = get_tickers('ibov')
ibov_tickers
```

```
ibov_tickers_retrive = yf.download(ibov_tickers, period='10y', group_by='tickers')
ibov_tickers_retrive.head
```

```
ifix = yf.download(get_tickers('ifix'), period='10y', group_by='tickers')
ifix.head
```

```
ibov_tickers_retrive.to_excel('ibov.xlsx')
files.download('ibov.xlsx')
```

```
ifix.to_excel('ifix.xlsx')
files.download('ifix.xlsx')
```

```
ibov.to_csv('ibov.csv')
files.download('ibov.csv')
```

```
ifix.to_csv('ifix.csv')
files.download('ifix.csv')
```
